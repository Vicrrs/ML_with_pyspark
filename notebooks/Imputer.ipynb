```python
spark = SparkSession.builder.appName("imputer").getOrCreate()
```

```python
from pyspark.ml.feature import Imputer
```

```python
carros = spark.read.csv("/content/drive/MyDrive/Colab Notebooks/Spark/datas/CarrosNAN.csv", header=True, inferSchema=True, sep=";")
carros.show(5)
```

```python
input = Imputer(inputCols=["Cilindradas", "Peso"], outputCols=["nova_Cilindradas", "novo_Peso"])
modelo = input.fit(carros)
carros = modelo.transform(carros)
carros.select("nova_Cilindradas", "novo_Peso").show()
```

```python
input = Imputer(inputCols=["Cilindros"], outputCols=["novo_Cilindros"])
input.setStrategy("median").setMissingValue(0).fit(carros).transform(carros).select("Cilindros", "novo_Cilindros").show()
```
