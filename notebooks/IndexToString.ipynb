```python
import pyspark
from pyspark.sql import SparkSession
```

```python
import findspark
findspark.init()
```

```python
# Conectando ao google colab
from google.colab import drive
drive.mount('/content/drive')
```

```python
spark = SparkSession.builder.appName("indextostring").getOrCreate()
```

```python
from pyspark.ml.feature import StringIndexer
```

```python
churn = spark.read.csv("/content/drive/MyDrive/Colab Notebooks/Spark/datas/Churn.csv", sep=";", inferSchema=True, header=True)
```

```python
indice = StringIndexer(inputCol="Geography", outputCol="indice")
modelo = indice.fit(churn)
dados_com_indice = modelo.transform(churn)
dados_com_indice.select("Geography", "indice").show(10)
```

## Convertendo valor numerico para categorico novamente, usando IndexToString

```python
from pyspark.ml.feature import IndexToString
```

```python
retorno = IndexToString(inputCol="indice", outputCol="categoriaoriginal")
```

```python
convertido = retorno.transform(dados_com_indice)
convertido.select("Geography", "indice", "categoriaoriginal").show(5)
```
