```python
import pyspark
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("binarizacao").getOrCreate()
```

```python
from pyspark.ml.feature import Binarizer
```

```python
iris = spark.read.csv("/content/drive/MyDrive/Colab Notebooks/Spark/datas/iris.csv", header=True, inferSchema=True)
```

```python
iris.show(5)
```

```python
binarizer = Binarizer(threshold=5, inputCol="sepallength", outputCol="sepallength_bin")
```

```python
irisbin = binarizer.transform(iris)
irisbin.show()
```
